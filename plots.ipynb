```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'model': ['Omics Baseline', 'Omics Baseline', 'CT Baseline','CT Baseline', 'CNN Simple Transforms', 'CNN Simple Transforms', 'CNN Augmented Image', 'CNN Agumented Image', 'CNN+Omics', 'CNN+Omics', 'CNN+Omics CV (AVG)', 'CNN+Omics CV (AVG)'], 
        'class':['0', '1']*6, 
        'precision' :[0.53, 0.57, 0.55, 0.64, 0.50, 0.17, 0.54, 0.33, 0.83, 1.00, 0.7786, 0.9286], 
        'recall':[0.60, 0.59, 0.73, 0.44, 0.58, 0.12, 0.64, 0.25, 1.00, 0.75, 0.9286, 0.6786],
        'f1': [0.56, 0.53, 0.63, 0.52, 0.54, 0.14, 0.58, 0.29, 0.91, 0.86, 0.8903, 0.7626]}

df = pd.DataFrame(data)
metrics = ['precision', 'recall', 'f1']
for metric in metrics:
    plt.figure(figsize=(12, 6))
    for label in ['0', '1']:
        subset = df[df['class'] == label]
        plt.plot(subset['model'], subset[metric], marker='o', label=f'Class {label}')
    plt.ylabel(metric)
    plt.title(f'{metric.capitalize()} by Model')
    plt.xticks(rotation=45, ha='right')
    plt.ylim(0, 1)
    plt.legend()
    plt.tight_layout()

    plt.show()


accuracy = [0.54, 0.58, 0.40, 0.89, 0.81]
models = ['Omics Baseline', 'CT Baseline, CNN Simple Transforms', 'CNN Augmented Image', 'CNN+Omics', 'CNN+Omics CV (AVG)']
plt.figure(figsize=(12, 6))
plt.plot(models, accuracy, marker='o')
plt.ylabel('Accuracy')
plt.title('Accuracy by Model')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.ylim(0, 1)
plt.show()
```
